```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("data_for_lr.csv")
```

```python
data.head()
```

```python
data.shape[0]
```

```python
data.shape
```

```python
data.columns
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
train_input=np.array(data["x"][:500],ndmin=2).reshape(500,1)
train_output=np.array(data["y"][:500],ndmin=2).reshape(500,1)

test_input=np.array(data["x"][500:],ndmin=2).reshape(199,1)
test_output=np.array(data["y"][500:],ndmin=2).reshape(199,1)
```

```python
train_input.shape
```

```python
test_input.shape
```

```python
model=LinearRegression()
```

```python
model.fit(train_input,train_output)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
predictions=model.predict(train_input)
```

```python
for i in range(train_input.shape[0]):
    print(f"Actual output = {train_output[i]}, prdicted_output = {predictions[i]}")
```

```python
from sklearn.metrics import mean_squared_error
```

```python
cost=mean_squared_error(predictions,train_output)
```

```python
cost
```

```python
plt.figure(figsize=(10,7))
plt.plot(train_input,train_output,"+",label="Train input",color="red")
plt.plot(train_input,predictions,"x",label="Predicted_output",color="green")
plt.title("LINEAR REGRESSION TRAIN MODEL")
plt.xlabel("Train input")
plt.ylabel("Train output & predictions")
plt.legend()
plt.show()
```

```python
testing=model.predict(test_input)
```

```python
for i in range(test_input.shape[0]):
    print(f"test_input ={test_input[i]} , predicted={testing[i]}")
```

```python
cost_function=mean_squared_error(testing,test_input)
```

```python
cost_function
```

```python
plt.figure(figsize=(10,7))
plt.plot(test_input,test_output,"*",label="test_output",color="blue")
plt.plot(test_input,testing,"x",label="predicted_output",color="red")
plt.title("Linear Regression Test Mode")
plt.xlabel("test input")
plt.ylabel("test output & predicted output")
plt.legend()
plt.show()
```

```python
model_predictions=np.append(predictions,testing)
```

```python
model_predictions
```

```python
model_input=np.append(train_input,test_input)
```

```python
for i in range(model_input.shape[0]):
    print(f"model input = {model_input[i]},model predicted = {model_predictions[i]}")
```
